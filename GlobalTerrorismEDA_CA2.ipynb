```python
from google.colab import drive
drive.mount('/drive')

#import libs
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
import seaborn as sns

plt.rcParams["figure.figsize"] = (28,9)
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.linewidth"] = 2.50

#reading the dataset
file = '/drive/MyDrive/globalterrorismdb_0718dist.csv'
data = pd.read_csv(file, encoding='ISO-8859-1')
```

```python
pd.set_option('max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
data.head()
```

```python
data.rename(columns={'iyear':'Year','extended':'Incident_Duration','city':'City','country_txt':'Country','region_txt':'Region','summary':'Summary', 'success':'Success',
                     'alternative_txt':'Attack_Tactic','attacktype1_txt':'Attack_type','suicide':'Suicide','targtype1_txt':'Target_type','targsubtype1_txt':'Target_SubCat',
                     'gname':'Terror_Group','motive':'Terror_Motive','weaptype1_txt':'Weapon_Type','weapsubtype1_txt':'Weapon_Subtype','nkill':'Total_Killed',
                     'nwoundus':'Total_Wounded','nkillter':'Terrorists_Killed','propextent_txt':'Prorperty_Damage','ishostkid':'Hostage_con',
                     'ransom':'Ransom','ransomamt':'Ransom_amt','ransompaid':'Ransom_amt_paid','hostkidoutcome_txt':'Hostage_outcome'}, inplace=True)
```

```python
finData = data[['Year', 'Incident_Duration', 'City', 'Country', 'Region', 'Summery', 'Success','Attack_Tactic','Attack_type',
                'Suicide','Target_type','Target_SubCat','Terror_Group','Terror_Motive','Weapon_Type','Weapon_Subtype','Total_Killed','Total_Wounded',
                'Terrorists_Killed','Prorperty_Damage','Hostage_con','Ransom','Ransom_amt','Ransom_amt_paid','Hostage_outcome']].reset_index(drop=True)
```

```python
finData.head()
```

```python
finData.isnull().sum()
```

```python
# City Column
finData['City'].fillna('Unknown', inplace = True)

# Summery Column 
finData['Summery'].fillna('Unknown', inplace = True)

# Terror_Motive column
finData['Terror_Motive'].fillna('Unknown', inplace = True)

# Prorperty_Damage column
finData['Prorperty_Damage'].fillna('Unknown', inplace = True)

# Attack_Tactic column
finData['Attack_Tactic'].fillna('Unknown', inplace = True)

# Target_SubCat column
finData['Target_SubCat'].fillna('Unknown', inplace = True)

# Weapon_Subtype column
finData['Weapon_Subtype'].fillna('Unknown', inplace = True)

# Filling null values in Hostage_outcome column
finData['Hostage_outcome'].fillna('Unknown', inplace = True)

# Summery column
finData['Summery'].fillna('Undocumented', inplace = True)

# Filling null values in Ransom_amt_paid column
finData['Ransom_amt_paid'].fillna(0, inplace = True)

# Filling null values in Ransom_amt column
finData['Ransom_amt'].fillna(0, inplace = True)

# Filling null values in Ransom_amt column
finData['Ransom'].fillna(-9.0, inplace = True)

# Replacing floating values with for better meaningful categorial value
finData[['Hostage_con','Ransom']] = finData[['Hostage_con','Ransom']].replace({0.0:'NO', 1.0:'Yes', -9.0:'Unknown'})
```

```python
# Total_Killed Column
finData['Total_Killed'] = finData['Total_Killed'].fillna(finData['Total_Killed'].mode()[0])

# Total_Wounded Column
finData['Total_Wounded'] = finData['Total_Wounded'].fillna(finData['Total_Wounded'].mode()[0])

# Terrorists_Killed Column
finData['Terrorists_Killed'] = finData['Terrorists_Killed'].fillna(finData['Terrorists_Killed'].mode()[0])

# Creating a new Column to show total_Casualities which is just a total number of people who got wounded and 
# killed in that incident including all Attackers, Defenders and Civilians

finData['Total_Casualities'] = finData['Total_Killed'] + finData['Total_Wounded']
```
